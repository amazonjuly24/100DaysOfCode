```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.hindustantimes.com/rss/topnews/rssfeed.xml'
```

```python
resp=requests.get(url)
```

```python
soup=BeautifulSoup(resp.content, features='xml')
```

```python
print(soup.prettify())
```

```python
items=soup.findAll('item')
```

```python
len(items)
```

```python
item=items[0]
```
